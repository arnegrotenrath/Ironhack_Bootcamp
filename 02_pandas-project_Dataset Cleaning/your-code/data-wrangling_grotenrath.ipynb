# 3 Check for empty and not-useful columns

```python
(sha_clean.shape[0]-sha_clean.count()) # Number of NaNs per column
```

Result: The Columns "Unnamed: 22" and "Unnamed: 23" can be dropped since they are almost completely empty. Age, Time and Species are just half filled, but the information in these columns can be valuable, hence they are kept

## 3.1 Dropping empty columns

```python
sha_clean.drop(["Unnamed: 22", "Unnamed: 23"], axis = 1, inplace = True)
```

## 3.2 Dropping non-useful columns

### 3.2.1 The column "original order" just counts the number of incidents. Since the incidents are ordered by date, this column contains the same value as the index of the row and can be dropped

```python
sha_clean["original order"].tail(20)
```

```python
sha_clean.drop(["original order"], axis = 1, inplace = True)
```

### 3.2.2 Case Number.1 and Case Number.2

A first look tells, that the columns "Case Number.1" and "Case Number.2" contains very often the exact same information compared as the column "Case Number":

```python
sha_clean[["Case Number", "Case Number.1", "Case Number.2" ]].head(10)
```

```python
sha_clean_case_number_equal1 = sha_clean["Case Number"] != sha_clean["Case Number.1"]
sha_clean_case_number_equal2 = sha_clean["Case Number"] != sha_clean["Case Number.2"]
```

```python
sha_clean[sha_clean_case_number_equal1].shape[0]# 
```

in Just 13 rows, "Case Number" and "Case Number.1" differ => the column can be dropped

```python
sha_clean.drop(["Case Number.1"], axis = 1, inplace = True)
```

```python
sha_clean[sha_clean_case_number_equal2].shape[0]
```

in Just 2 rows, "Case Number" and "Case Number.2" differ => the column can be dropped

```python
sha_clean.drop(["Case Number.2"], axis = 1, inplace = True)
```

### 3.2.3 Additional, but not evaluable information in columns "pdf", "href formula", "href" and "name"

Since only links to additional information is provided in these 3 columns, they can be dropped as well. The useful information isn't lost, since we drop these columns only in the clean DataFrame, which is a copy of the original one

```python
sha_clean.drop(["pdf", "href formula", "href", "Name"], axis = 1, inplace = True)
```

# 4 Renaming columns: "Sex " to "Sex", "Species " to "Species"

```python
sha_clean.rename(columns = {"Sex ": "Sex", "Species ": "Species"}, inplace = True)
```

```python
sha_clean[["Sex"]].groupby("Sex").size()
```

# 5 Aggregating associated values

## 5.1 For column "Type"

```python
sha_clean[["Type"]].groupby("Type").size()
```

The Types Boat and Boating can be combined

```python
sha_clean["Type"].where(sha_clean["Type"] != "Boating", "Boat", inplace = True)
```

```python
sha_clean[["Type"]].groupby("Type").size()
```

## 5.2 For column "Fatal"

```python
sha_clean[["Fatal (Y/N)"]].groupby("Fatal (Y/N)").size()
```

```python
rename_list_n =["F", "n", " N", "N "]
for element in rename_list_n:
    sha_clean["Fatal (Y/N)"].where(sha_clean["Fatal (Y/N)"] != element, "N", inplace = True)
    
sha_clean["Fatal (Y/N)"].where(sha_clean["Fatal (Y/N)"] != "#VALUE!", "UNKNOWN", inplace = True)
```

```python
sha_clean[["Fatal (Y/N)"]].groupby("Fatal (Y/N)").size()
```

```python
sha_clean.head(10)
```

## 5.3 Compare Case Number, Date and Year

```python
sha_clean.head(10)
```

The columns Case Number, Date and Year contain similar information, if Case Number is cleaned up.

```python
sha_clean["Case Number"] = sha_clean["Case Number"].str.replace("\D*\Z", "", regex = True) # deleting the character for multiple events on the same day
sha_clean["Case Number"] = sha_clean["Case Number"].str.replace("(N|n).*", "", regex = True)# deleting non-date events
sha_clean["Case Number"] = sha_clean["Case Number"].str.replace(".*?(?=\d{4}\Z)", "", regex = True)
```

```python
sha_clean.tail(10)
```

Replace the due to unreadable values empty rows in column Case Number with nan

```python
sha_clean[["Case Number"]] = sha_clean[["Case Number"]].replace(r"^\s*$", np.nan, regex = True)
```

Now, the columns Date and Year can be dropped and "Case Number" can be renamed into date, since it contains exactly that

```python
sha_clean.drop(["Date", "Year"], axis = 1, inplace = True)
```

```python
sha_clean.rename(columns = {"Case Number": "Date"}, inplace = True)
```
