```python
df_öpnv_original = pd.read_csv("200130_ÖPNV_Statistik_Bundesländer_clean1_46100-0011.csv", encoding = "ISO-8859-1",sep = ";")
df_öpnv = pd.read_csv("200130_ÖPNV_Statistik_Bundesländer_clean1_46100-0011.csv", encoding = "ISO-8859-1",sep = ";")
```

Due to unreadable and unnecessary values, the first and last three columns are dropped

```python
df_öpnv.drop([0, df_öpnv.shape[0]-3 ,df_öpnv.shape[0]-2, df_öpnv.shape[0]-1], axis = 0, inplace = True)
```

In order to keep the evaluation managable, a few columns are dropped, since they are not our main focus

```python
df_öpnv.drop(["Unternehmensart", "Unternehmen", "Beförderungsangebot", "Einnahmen im Liniennahverkehr"], axis = 1, inplace = True)
```

Change type of numeric columns to numeric type

```python
df_öpnv = df_öpnv.astype({"Jahr": "int32",
                          "Beförderte Personen": "int64",
                          "Beförderungsleistung": "int64",
                          "Fahrleistung": "int64"})
```

Aggregate for Year and state

```python
df_öpnv = df_öpnv.groupby(["Jahr", "Bundesland"]).sum().reset_index()
```

```python
df_pop = pd.read_csv("200201_Bevölkerungszahlen_clean1_12211-0005.csv", encoding = "ISO-8859-1",sep = ";")
```

```python
df_pop.tail()
```

drop the last four rows since they don't contain the desired values

```python
df_pop.drop([df_pop.shape[0]-4, df_pop.shape[0]-3, df_pop.shape[0]-2, df_pop.shape[0]-1], inplace = True)
```

Cleaning the year values so that only the year is displayed

```python
df_pop["Jahr"] = df_pop[["Jahr"]].replace("\d{2}/", "", regex = True)
```

```python
df_pop = df_pop.astype({"Jahr": "int64",
                       "Bevölkerung": "int64",
                       "Erwerbstätige": "int64",
                       "Erwerbslose": "int64",
                       "Erwerbspersonen": "int64",
                       "Nichterwerbspersonen": "int64"})
```

For now, we are only interested in the population of the state, stored in the column "Bevölkerung"

```python
df_pop.drop(["Erwerbstätige", "Erwerbslose", "Erwerbspersonen", "Nichterwerbspersonen"], axis = 1, inplace = True)
```

The unit of the original dataset is 1000, so we multiply with that

```python
df_pop["Bevölkerung"] = df_pop["Bevölkerung"]*1000
```

Drop all rows with years before 2004, since we only have train data for these years

```python
index_years = df_pop[df_pop["Jahr"] < 2004].index
```

```python
df_pop.drop(index_years, inplace = True)
```

```python
df_öpnv_pop = pd.merge(df_öpnv, df_pop, how="left", left_on=["Jahr", "Bundesland"], right_on = ["Jahr", "Bundesland"])
```

Create a dataframe with the area of every state since the api of the german government is down

```python
states = df_öpnv["Bundesland"].unique()
```

```python
area_states = [35674, 70542, 891, 29654, 420, 755, 21116, 23293, 47710, 34113, 19858, 2571, 18450, 20452, 15802, 16202]
```

```python
df_area = pd.DataFrame({"Bundesland": states,
                         "Fläche": area_states})
```

```python
df_area
```

```python
df_öpnv_pop_area = pd.merge(df_öpnv_pop, df_area, how="left", left_on="Bundesland", right_on ="Bundesland")
```

```python
df_öpnv_pop_area["Bevölkerungsdichte"] = df_öpnv_pop_area["Bevölkerung"] / df_öpnv_pop_area["Fläche"]
```

```python
df_öpnv_pop_area.to_csv("Bahnstatistik.csv", index = False)
```
